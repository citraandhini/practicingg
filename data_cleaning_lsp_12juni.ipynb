```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
#membaca file dataset dengan dataFrame atau df
df = pd.read_csv('/content/Popular_Spotify_Songs.csv', encoding="ISO-8859-1")
```
