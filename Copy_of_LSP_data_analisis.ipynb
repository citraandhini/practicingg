```python
netflix.isnull()
```

```python
#to show the count of no value (angka kosong)
miss=netflix.isnull().sum()
miss
```

```python
#for count in percent
miss1 = netflix.isnull().sum()/len(netflix)* 100
miss1
```

```python
#testing values with %
m = pd.concat([miss,miss1],axis=1,keys=['Total','missing%'])
```

```python
#using heatmap to shows that null values count
sns.heatmap(netflix.isnull())
```

```python
#making copy of dataset for changes
netflix_copy = netflix.copy()
netflix_copy.head()
```

```python
#for delete all rows that contai same values or null
netflix_copy = netflix_copy.dropna(how='any',subset=['director','cast'])
netflix_copy.head(2)
```

```python
netflix_copy.fillna({'country':'missing','rating':'missing','duration':'missing'},inplace=True)
```
